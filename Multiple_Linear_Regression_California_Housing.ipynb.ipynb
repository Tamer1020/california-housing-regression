```python
X = df.iloc[:, :-1]   
y = df.iloc[:, -1]    
print(X.head())

```

```python
print(y.head())
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.33,
    random_state=10
)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

## Modell Training

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()

model.fit(X_train_scaled, y_train)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
y_pred = model.predict(X_test_scaled)
y_pred[:10]
```

```python
y_test.values[:10]
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

mse  = mean_squared_error(y_test, y_pred)
mae  = mean_absolute_error(y_test, y_pred)
rmse = mse ** 0.5

r2 = r2_score(y_test, y_pred)

n = X_test.shape[0]     
p = X_test.shape[1]     
adjusted_r2 = 1 - ((1 - r2) * (n - 1) / (n - p - 1))

mse, mae, rmse, r2, adjusted_r2
```

## Residual Plot

```python
residuals = y_test - y_pred

plt.figure(figsize=(8, 6))

sns.scatterplot( x=y_pred , y=residuals ,alpha =0.5)
plt.axhline(y=0, color='red', linestyle='--')
plt.xlabel("Predicted Prices")
plt.ylabel("Residuals (y_true - y_pred)")
plt.title("Residual Plot: Predicted Prices vs Residuals")
plt.show()
```

```python
# Residual Plot Analysis
#
# The residual plot shows that the errors are not randomly distributed around zero,
# which indicates that several key assumptions of linear regression are violated.
#
# 1. Heteroscedasticity:
#    The spread of residuals increases as the predicted house price increases.
#    This triangular or fan-shaped pattern suggests that the model performs
#    better on lower-priced houses and worse on higher-priced ones.
#
# 2. Non-linearity:
#    The clear downward trend in the residuals indicates that the relationship
#    between the features and the target variable is not perfectly linear.
#    A linear regression model cannot fully capture these patterns.
#
# 3. Outliers:
#    Several large negative residuals (e.g., -6 to -10) indicate that the model
#    significantly overestimated the price for some houses.
#
# Overall, the residual plot suggests that a simple linear regression model is not
# sufficient for this dataset. Regularization techniques such as Ridge, Lasso, or
# ElasticNetor even non-linear modelsmay provide better performance.
```

## Distribution of Residuals

```python
plt.figure(figsize=(8, 5))

sns.histplot(residuals , kde=True, bins=40, color='blue' )
plt.axvline(x=0, color='red', linestyle='--')
plt.title("Distribution of Residuals")
plt.xlabel("Residuals")
plt.ylabel("Frequency")


plt.show()
```

```python
# Distribution of Residuals Analysis
#
# The residuals are centered around zero, indicating low overall bias.
# However, the distribution is clearly left-skewed, with a long negative tail
# extending down to -10. This suggests that the model significantly overestimates
# the price for some samples (especially higher-priced houses).
#
# The lack of symmetry in the residual distribution indicates violations of the
# normality assumption of linear regression. Additionally, the presence of strong
# negative outliers increases error metrics such as MSE and RMSE.
#
# Overall, the residual distribution suggests that simple linear regression does
# not fully capture the underlying structure of the data. Regularization methods
# (Ridge, Lasso, ElasticNet) or non-linear models may provide more robust results.
```

## Ridge Regression

```python
from sklearn.linear_model import Ridge

alphas = [0.001, 0.01, 0.1, 1, 10, 50, 100, 200, 500]
results = []
for a in alphas:
    ridge = Ridge(alpha=a)
    ridge.fit(X_train_scaled, y_train)
    y_pred_r = ridge.predict(X_test_scaled)

    mse  = mean_squared_error(y_test, y_pred_r)
    mae  = mean_absolute_error(y_test, y_pred_r)
    rmse = mse ** 0.5
    r2   = r2_score(y_test, y_pred_r)

    
    results.append([a, mse, mae, rmse, r2])

df_results = pd.DataFrame(results, columns=["alpha", "MSE", "MAE", "RMSE", "R2"])
df_results
    
```

```python
# Ridge Regression Performance Summary
# Looping over multiple alpha values shows how regularization strength affects model accuracy.
# 
# 1) Small alpha values (0.0010.1):
#     Very similar to Linear Regression
#     Regularization effect is minimal
#
# 2) Medium alpha values (10200):
#     Best performance occurs here
#     alpha = 100 gives the highest R ( 0.59515) 
#      and one of the lowest RMSE values ( 0.7417)
#     Indicates that moderate regularization helps stabilize the model
#
# 3) Large alpha values (500+):
#     Performance drops (higher RMSE, lower R)
#     Model becomes too restricted  underfitting
#
# Conclusion:
# Ridge Regression slightly improves model stability compared to simple Linear Regression.
# The optimal  is around 100, balancing bias and variance effectively.
```

## Alpha vs R Plot

```python


plt.figure(figsize=(10,5))
plt.plot(df_results["alpha"], df_results["R2"], marker="o", linestyle="-", color="blue")

plt.xscale("log")  #    alpha    
plt.xlabel("Alpha (log scale)")
plt.ylabel("R Score")
plt.title("Ridge Regression: Alpha vs R Score")
plt.grid(True)
plt.show()

```

```python
plt.figure(figsize=(10,5))
plt.plot(df_results["alpha"], df_results["RMSE"], marker="o", linestyle="-", color="red")

plt.xscale("log")
plt.xlabel("Alpha (log scale)")
plt.ylabel("RMSE")
plt.title("Ridge Regression: Alpha vs RMSE")
plt.grid(True)
plt.show()
```

## Lasso Regression

```python
from sklearn.linear_model import Lasso

alphas = [0.001, 0.01, 0.1, 1, 10, 50, 100, 200, 500]
results_lasso = []
for alpha in alphas:
    lasso = Lasso(alpha=alpha, max_iter=5000)
    lasso.fit(X_train_scaled, y_train)
    y_pred = lasso.predict(X_test_scaled)

    mse  = mean_squared_error(y_test, y_pred)
    mae  = mean_absolute_error(y_test, y_pred)
    rmse = mse ** 0.5
    r2   = r2_score(y_test, y_pred)

    results_lasso.append([alpha, mse, mae, rmse, r2])

df_lasso = pd.DataFrame(results_lasso, columns=["alpha", "MSE", "MAE", "RMSE", "R2"])
df_lasso
```

## ElasticNet Regression

```python
from sklearn.linear_model import ElasticNet
alpha_values = [0.001, 0.01, 0.1, 1, 10, 50]

l1_values = [0.1, 0.3, 0.5, 0.7, 0.9]

results_elastic = []

for alpha in alpha_values:
    for l1 in l1_values:
        model = ElasticNet(alpha=alpha, l1_ratio=l1, max_iter=5000)
        model.fit(X_train_scaled, y_train)
        y_pred = model.predict(X_test_scaled)

        mse  = mean_squared_error(y_test, y_pred)
        mae  = mean_absolute_error(y_test, y_pred)
        rmse = mse ** 0.5
        r2   = r2_score(y_test, y_pred)

        results_elastic.append([alpha, l1, mse, mae, rmse, r2])

df_elastic = pd.DataFrame(results_elastic, 
                           columns=["alpha", "l1_ratio", "MSE", "MAE", "RMSE", "R2"])

df_elastic
```

```python
#ElasticNet was evaluated over a grid of alpha and l1_ratio values. 
#The best performance is obtained for alpha = 0.01 and l1_ratio = 0.7, 
#with the lowest RMSE (~0.7409) and the highest R (~0.5961). 
#This setting provides a good balance between L1 (feature selection) 
#and L2 (coefficient shrinkage) regularization.

#For very small alpha values (0.001), the model behaves similarly to 
#Ridge/Lasso with mild regularization, achieving R around 0.594. 
#As alpha increases to 0.1 and beyond, both RMSE increases and R decreases, 
#indicating that the model becomes over-regularized and starts to underfit.

#Overall, ElasticNet slightly outperforms both Ridge and Lasso on this dataset, 
#showing that a combination of L1 and L2 penalties (l1_ratio  0.7) 
#is more effective than using pure L1 or pure L2 regularization alone.
```

```python
pivot_r2 = df_elastic.pivot(index="l1_ratio", columns="alpha", values="R2")

plt.figure(figsize=(10, 6))
sns.heatmap(pivot_r2, annot=True, cmap="viridis", fmt=".4f")
plt.title("ElasticNet Heatmap: R Score")
plt.show()

```

```python
# ElasticNet R Heatmap Analysis

# The heatmap shows how the models performance (R score) changes across
# different combinations of alpha (regularization strength) and l1_ratio 
# (balance between L1 and L2 penalties).

# Best Performance:
#   The highest R (~0.596) occurs when alpha is small (0.01) and l1_ratio is in 
#   the range 0.30.7. This indicates that a balanced mix of L1 and L2 
#   regularization works best for this dataset.

#  Effect of alpha:
#   Very small alpha (0.001) gives stable but slightly lower performance.
#   Moderate alpha (0.01) is the clear optimum.
#   Large alpha values ( 10) cause severe underfitting, dropping R to ~0.

#  Effect of l1_ratio:
#   Increasing l1_ratio up to 0.7 improves performance slightly.
#   Very high l1_ratio values make the model too sparse and reduce performance.

#  Conclusion:
#   ElasticNet outperforms pure Lasso or Ridge on this dataset.
#   The dataset benefits from mild regularization and a balanced mix of L1 and L2.
```

### Final Model Training (Best ElasticNet)

```python
# Final full dataset for retraining the best model
X_full = df.drop("price", axis=1)
y_full = df["price"]
final_scaler = StandardScaler()
X_full_scaled = final_scaler.fit_transform(X_full)
```

```python
best_model = ElasticNet(alpha=0.01, l1_ratio=0.7, max_iter=5000)
best_model.fit(X_full_scaled, y_full)
```
